```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the House_Price Prediction dataset
df = pd.read_csv('./Datasets/House Data.csv')

# Compute standard deviation for each feature
std_dev_values = df.std()
print("\nStandard deviation for each feature:")
print(std_dev_values)

# Compute variance for each feature
variance_values = df.var()
print("\nVariance for each feature:")
print(variance_values)

# Compute percentiles (25th, 50th, and 75th) for each feature
percentiles_values = df.describe(percentiles=[.25, .5, .75]).loc[['25%', '50%', '75%']]
print("\nPercentiles for each feature:")
print(percentiles_values)

# Create histograms for each feature
for column in df.columns:
    plt.figure(figsize=(8, 5))
    plt.hist(df[column], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram for {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```
