(Implement step by step using commands - Dont use library) Use this dataset to build a decision tree, with Buys as the target variable, to help in buying lipsticks in the future. Find the root node of the decision tree.

```python
import pandas as pd

df = pd.read_csv("./Datasets/Lipstick.csv")

```

```python
class TreeNode:
    def __init__(self, data=None, attribute=None, value=None, outcome=None):
        self.data = data
        self.attribute = attribute  # Feature to split on
        self.value = value  # Value of the feature to make the split
        self.outcome = outcome  # Class label for leaf nodes
        self.children = {}

    def add_child(self, value, node):
        self.children[value] = node

def get_majority_class(labels):
    # Return the majority class in a set of labels
    return max(set(labels), key=labels.count)

def build_decision_tree(data):
    # Recursive function to build the decision tree

    # If all examples have the same class, create a leaf node
    if len(set(data['Buys'])) == 1:
        return TreeNode(data=data, outcome=data['Buys'].iloc[0])

    if len(data.columns) == 1:
        return TreeNode(data=data, outcome=get_majority_class(data['Buys']))
    best_attribute, best_value = find_best_split(data)
    subsets = split_data(data, best_attribute, best_value)
    node = TreeNode(data=data, attribute=best_attribute, value=best_value)

    for subset_value, subset_data in subsets.items():
        child_node = build_decision_tree(subset_data)
        node.add_child(subset_value, child_node)

    return node

def find_best_split(data):
    return data.columns[1], data[data.columns[1]].iloc[0]

def split_data(data, attribute, value):
    return {'subset1': data.iloc[:len(data)//2], 'subset2': data.iloc[len(data)//2:]}

decision_tree = build_decision_tree(df)
```
