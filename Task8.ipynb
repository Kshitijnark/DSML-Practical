8. Write a program to do: A dataset collected in a cosmetics shop showing details of customers and whether or not they responded to a special offer to buy a new lip-stick is shown in table below. (Use library commands) According to the decision tree you have made from the previous training data set, what is the decision for the test data: [Age = 21-35, Income = Low, Gender = Male, Marital Status = Married]?

```python
import numpy as np
import pandas as pd

print("Loading the required libaries...")
```

```python
df = pd.read_csv("./Datasets/Lipstick.csv")
df
```

**Describing the data**

```python
df.describe()
```

```python
df['Buys'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
#data=data.apply(le.fit_transform)
x=df.iloc[:,:-1] #-1 means don't take last column 
x=df.drop(columns='Id')
x=df.drop(columns='Buys')
x=x.apply(le.fit_transform)
#find label with their encoded value
print("Age with encodd value :",list( zip(df.iloc[:,0], x.iloc[:,0])))
print("\nIncome with encoded value :",list( zip(df.iloc[:,1], x.iloc[:,1])))
print("\nGender with encoded value :",list( zip(df.iloc[:,2], x.iloc[:,2])))
print("\nmaritialStatus with encoded value :",list( zip(df.iloc[:,3], x.iloc[:,3])))
```

```python
df
```

```python
x.drop(columns='Id',axis=1,inplace=True)
x
```

```python
#Store labels in Y
y=df.iloc[:,-1]
y
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion='entropy')
classifier.fit(x.values,y.values)
```

```python
#  Q5
#  [Age < 21, Income = Low, Gender = Female, Marital Status = Married]

test_x=np.array([1,1,0,0])
pred_y=classifier.predict([test_x])
pred_y[0]
```

```python
# Q8 [Age = 21-35, Income = Low,Gender = Male, Marital Status = Married]

test_x=np.array([0,2,1,0])
pred_y=classifier.predict([test_x])
pred_y[0]
```
