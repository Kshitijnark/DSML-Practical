## Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn

print("Done with imports!")
```

## Starting the analysis

```python
df = pd.read_csv('Iris.csv')

df
```

```python
df.describe()
```

## Creating a box plot for each feature!

```python
plt.title("Box plot for Petal Length")

sn.boxplot(data=df['petallength'])

plt.show()
```

```python
plt.title("Box plot for Petal Width")

sn.boxplot(data=df['petalwidth'])

plt.show()
```

```python
plt.title("Box plot for Sepal Length")

sn.boxplot(data=df['sepallength'])

plt.show()
```

```python
plt.title("Box plot for Sepal Width")

sn.boxplot(data=df['sepalwidth'])

plt.show()

# Outliners can be as dot or diamond shaped point is box plot
```
