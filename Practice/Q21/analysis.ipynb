```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import seaborn
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('iris_csv.csv')

df
```

# Building our model

```python
Y = df['class']
X = df.drop(columns=['class'])
```

```python
clustering = KMeans(n_clusters=4,random_state=3,max_iter=10)

clustering.fit(X)
```

# Plotting the graph

```python
colorArr = np.array(['g','r','b','m'])

plt.scatter(x=X.sepallength,y=X.sepalwidth,c=colorArr[clustering.labels_])

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")

plt.show()
```

```python
colorArr = np.array(['g','r','b','m'])

plt.scatter(x=X.petallength,y=X.petalwidth,c=colorArr[clustering.labels_])

plt.xlabel("Petal Length")
plt.ylabel("Petal Width")

plt.show()
```
