# Box plot

```python
species = df['class']
sepal_length = df['sepallength']
petal_length = df['petallength']
petal_width = df['petalwidth']
sepal_width = df['sepalwidth']
```

```python
sn.boxplot(x=species,y=sepal_length)
plt.show()
```

```python
sn.boxplot(x=species,y=sepal_width)
plt.show()
```

```python
sn.boxplot(x=species,y=petal_length)
plt.show()
```

```python
sn.boxplot(x=species,y=petal_width)
plt.show()
```

# Histogram

```python
sn.histplot(x=sepal_length,data=df)
plt.show()
```

```python
sn.histplot(x=sepal_width,data=df)
plt.show()
```
