## Visualization

### Price vs bedrooms (Bar plot)

```python
plt.title("Price vs bedrooms")
sn.barplot(x=Df['bedrooms'],y=Df['price'])
plt.ylabel("Price (In Lakhs)")
plt.show()
```

## Box plot

```python
plt.title("Bathrooms boxplot")
sn.boxplot(data=Df['bathrooms'])
plt.show()
```

## Histogram

```python
sn.histplot(data=Df['sqft_above'])
plt.show()
```
