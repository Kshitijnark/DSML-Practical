```python
# Import necessary libraries
import pandas as pd
df = pd.read_csv("./Datasets/Titanic.csv")
print("First few rows of Titanic CSV dataset:")
print(df.head())

```

```python

selected_columns = ['Pclass', 'Name', 'Sex', 'Age', 'Survived']
selected_data_csv = df[selected_columns]
print("\nSelected data from Titanic CSV dataset:")
print(selected_data_csv.head())
```

```python


sorted_data_csv = df.sort_values(by='Age')

print("\nSorted data by Age in Titanic CSV dataset:")
print(sorted_data_csv.head())
```

```python

description_csv = df.describe()

print("\nAttribute description in Titanic CSV dataset:")
print(description_csv)

data_types_csv = df.dtypes

print("\nData types in Titanic CSV dataset:")
print(data_types_csv)
```
