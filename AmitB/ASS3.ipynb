```python
import pandas as pd
data=pd.read_csv("Datasets/House Data.csv")
data
```

```python
d=data.select_dtypes(exclude=[object])
std_val=d.std()
std_val
```

```python
var_val=d.var()
var_val
```

```python
perc=data.describe(percentiles=[0.25,0.5,0.75]).loc[['25%','50%','75%']]
perc
```

```python
col=data.columns.to_list()

print(col)
import matplotlib.pyplot as plt
for i in col:
    plt.hist(data[i],bins=10)
    # plt.title("Histogram of ",i)
    plt.xlabel(i)
    plt.ylabel("Frequency")
    plt.xticks(data[i])
    plt.show()
```
